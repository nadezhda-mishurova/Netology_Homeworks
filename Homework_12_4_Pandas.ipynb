К домашнему заданию, задача 4
Дана статистика услуг перевозок клиентов компании по типам:

rzd - железнодорожные перевозки
auto - автомобильные перевозки
air - воздушные перевозки
client_base - адреса клиентов

```python
import pandas as pd
```

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
#Попеременно сделаем сливание таблиц по принципу "ничего не потерять" в каждый новый датафрейм 
# В качестве индекса оставим client_id последнего датафрейма с адресами ( т.е. наиболее полного)
# В каждом датафрейме заполним NaN значения нулями 
# Начнем c РЖД, далее Авто и в конце - Авиа
clint_base_all = client_base.merge(rzd, how ='outer', on ='client_id').merge(auto, how ='outer', on ='client_id').merge(air, how ='outer', on ='client_id').fillna(0)
clint_base_all
```
